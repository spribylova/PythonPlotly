```python
import holoviews as hv
from holoviews import opts, dim
import holoviews.plotting.bokeh
from bokeh.sampledata.les_mis import data
import pandas as pd
from bokeh.plotting import show, output_file
import numpy as np
hv.extension('bokeh')
hv.output(size=200)
```

```python
data2 = [
        [0, 6703, 6746, 12342 ], # install
        [0, 594, 2972, 2701], # purchase
        [0, 1082, 2802, 169 ], # reopen
        [0, 2596, 3880, 16]] # signup
```

```python
data2
```

```python
# Declare holoviews grid
df2 = pd.DataFrame (data2, columns = ['install', 'purchase', 'reopen', 'signup'])
df2
```

```python
events = ['install', 'purchase', 'reopen', 'signup']
events
```

```python
data2 = hv.Dataset((events,events, df2), ['target', 'source'], 'value').dframe()

# Declare a gridded HoloViews dataset and call dframe to flatten it
# data = hv.Dataset((list(df.columns), list(df.index), df),['source', 'target'], 'value').dframe()
```

```python
data2
```

```python
# create the chord diagram from the transformed data
chord = hv.Chord(data2)
chord.opts(
    node_color='index', edge_color='source', label_index='index', 
    cmap='Greys', edge_cmap='Greys', width=200, height=200, 
    edge_hover_line_color=None,  node_selection_fill_color=None, edge_alpha= 0.8)
output_file('chordtest.html')
show(hv.render(chord))
```
