```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
import json
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns # optional, will only affect the color of bars and the grid
from ipywidgets import widgets, interactive
```

```python
# Marke,Mitarbeiter 2025,Gegründet,Stadt,Land,Filialen,Mitarbeiter per Filiale,CEO 2025,Website
df = pd.read_csv('DE_Mode.csv', delimiter=',', encoding="utf-8-sig")
df = df.sort_values('Mitarbeiter 2025', ascending=False)
```

```python
# animation frame
df2 = df.melt(id_vars='Marke', value_vars=df.columns, var_name='Mitarbeiter 2025')
```

```python
# 1. Plotly Static
# Employees in the year 2025 by Brand ( with Headquarters in Germany founded before 1950)
```

```python
land, gegrundet = "DE", 1950
df_land = df[ (df["Land"] == land) &  (df['Gegründet'] <= gegrundet) ]
fig = px.bar(df_land, x = 'Marke', y = 'Mitarbeiter 2025',  color ='Mitarbeiter 2025' , color_continuous_scale = 'Greys', title="Employees in the year 2025 by Brand ( with Headquarters in Germany founded before 1950) ")
fig.update_layout(xaxis={'categoryorder': 'total descending'})
fig.show()
```
